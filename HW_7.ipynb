Для проверки различий между двумя независимыми выборками, в данном случае, когда не соблюдается условие нормальности, можно использовать непараметрический критерий Манна-Уитни.


Выбрать тест и проверить, есть  ли различия между выборками:

1 )  Даны две  независимые выборки. Не соблюдается условие нормальности

x1  380,420, 290

y1 140,360,200,900

Сделайте вывод по результатам, полученным с помощью функции

```python
from scipy.stats import mannwhitneyu

x1 = [380, 420, 290]
y1 = [140, 360, 200, 900]

stat, p = mannwhitneyu(x1, y1)
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p > alpha:
    print('Принять гипотезу о равенстве выборок')
else:
    print('Отклонить гипотезу о равенстве выборок')

# Если p-value меньше заданного уровня значимости (например, 0.05), то мы можем отклонить гипотезу о равенстве выборок.
# Если p-value больше уровня значимости, то мы не можем отвергнуть гипотезу о равенстве выборок.
# Результаты теста Манна-Уитни могут указывать на наличие или отсутствие различий между выборками. 
# Но поскольку у нас нет информации о том, что представляют собой эти выборки, невозможно сделать вывод о том, как эти различия могут быть значимыми.
```

# 3
Сравните 1 и 2 е измерения, предполагая, что 3го измерения через 30 минут не было.

```python
from scipy.stats import wilcoxon

# данные измерений до препарата и через 10 минут
x1 = [150, 160, 165, 145, 155]
x2 = [140, 155, 150, 130, 135]

# проводим тест Уилкоксона
stat, p = wilcoxon(x1, x2)

# выводим результаты
print('Статистика = %.3f, p-значение = %.3f' % (stat, p))
if p > 0.05:
    print('Нет статистически значимых различий')
else:
    print('Есть статистически значимые различия')
```

# 4
 Даны 3 группы  учеников плавания.

В 1 группе время на дистанцию 50 м составляют:

56, 60, 62, 55, 71, 67, 59, 58, 64, 67

Вторая группа : 57, 58, 69, 48, 72, 70, 68, 71, 50, 53

Третья группа: 57, 67, 49, 48, 47, 55, 66, 51, 54

Проверить, есть ли различия между группами

```python
from scipy.stats import kruskal

# время на дистанцию 50 м для каждой группы
group1 = [56, 60, 62, 55, 71, 67, 59, 58, 64, 67]
group2 = [57, 58, 69, 48, 72, 70, 68, 71, 50, 53]
group3 = [57, 67, 49, 48, 47, 55, 66, 51, 54]

# применяем критерий Краскела-Уоллиса для проверки гипотезы о равенстве средних
stat, p = kruskal(group1, group2, group3)

# устанавливаем уровень значимости
alpha = 0.05

# выводим результаты
if p < alpha:
    print('Найдены статистически значимые различия между группами.')
else:
    print('Статистически значимых различий между группами не найдено.')
```

# 5
Заявляется, что партия изготавливается со средним арифметическим 2,5 см. Проверить данную гипотезу, если известно, что размеры изделий подчинены нормальному закону распределения. Объем выборки 10, уровень статистической значимости 5%

2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

спользуем одновыборочный t-тест для проверки гипотезы о равенстве среднего значения генеральной совокупности заданному значению 2.5.

В данном случае:

Генеральная совокупность - размеры изделий
Заданное среднее значение генеральной совокупности - 2.5
Выборка - 2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

```python
import scipy.stats as stats

# Заданный уровень статистической значимости
alpha = 0.05

# Выборочное среднее
x_bar = sum([2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34]) / 10

# Стандартное отклонение
s = ((2.51 - x_bar)**2 + (2.35 - x_bar)**2 + (2.74 - x_bar)**2 + (2.56 - x_bar)**2 + (2.40 - x_bar)**2 + (2.36 - x_bar)**2 + (2.65 - x_bar)**2 + (2.7 - x_bar)**2 + (2.67 - x_bar)**2 + (2.34 - x_bar)**2) / 9
s = s**0.5

# Количество степеней свободы
df = 9

# Вычисление t-статистики
t = (x_bar - 2.5) / (s / (10**0.5))

# Вычисление критической t-статистики
t_critical = stats.t.ppf(1 - alpha, df)

# Проверка гипотезы
if t < -t_critical or t > t_critical:
    print("Гипотеза отвергается")
else:
    print("Гипотеза не отвергается")
```

Вывод: так как полученное значение t-статистики не попадает в критическую область, то гипотеза о том, что партия изготавливается со средним арифметическим 2,5 см, не отвергается на уровне статистической значимости 5%.
